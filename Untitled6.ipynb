```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Function to load dataset from CSV file
def load_dataset(file_path):
    try:
        df = pd.read_csv("C:\\Users\\G.DEV AMARNADH\\Desktop\\Give-Life-Predict-Blood-Donations-master\\datasets\\transfusion.csv")
        return df
    except FileNotFoundError:
        print("File not found. Please make sure you provide the correct file path.")
        return None
    except Exception as e:
        print("An error occurred:", e)
        return None

# Load dataset
file_path = input("Enter the file path of the dataset (CSV format): ")
df = load_dataset(file_path)

if df is not None:
    # Displaying the DataFrame
    print("\nDataFrame:")
    print(df)

    # Summary statistics
    print("\nSummary Statistics:")
    print(df.describe())

    # Checking for missing values
    print("\nMissing Values:")
    print(df.isnull().sum())

    # Visualizing data
    # Pairplot
    sns.pairplot(df, hue='whether he/she donated blood in March 2007')
    plt.show()

    # Correlation matrix
    print("\nCorrelation Matrix:")
    print(df.corr())
```
